```python
import numpy as np
from scipy.linalg import svd
```

```python
class Exercise:
    def __init__(self, m, p, k, alpha=0.5):
        self.m = m
        self.p = p
        self.k = k
        self.alpha = alpha
         
    def solver(self):
        
        matrix = self.random_sampler()
        U, Sigma, V_T = self.SVD(matrix.reshape(self.m ,self.m))
        
        E = np.dot(U[:, 1:self.k], np.power(Sigma[1:self.k], self.alpha))

        print('Generated A :\n', matrix)
        print('='*50)
        print('U matrix:\n', U)
        print('='*50)
        print('Sigma matrix:\n', Sigma)
        print('='*50)
        print('V matrix:\n', V_T)
        print('='*50)
        print('E matrix:\n', E)
#         return E
    
    def random_sampler(self):
        
        random_matrix = np.zeros(self.m*self.m) 
        nonzeros_count = int(np.rint(self.m*self.m*(1-self.p)))
        idx = np.random.choice(self.m*self.m, nonzeros_count, replace=0)
        nonzeros_num = np.random.randint(1, 30, nonzeros_count)
        random_matrix[idx] = nonzeros_num
        
        return random_matrix
    
    def SVD(self, A):

        U, sigma, V_trans = svd(A)
        
        return U, sigma, V_trans
```

```python
m = 10
zero_ratio = 0.8
alpha = 0.5
k = 5

ex_1 = Exercise(m, zero_ratio, k, alpha)
ex_1.solver()
```
